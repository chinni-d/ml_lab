```python
import seaborn as sns
```

```python
data_set=sns.load_dataset("Iris")
data_set
```

```python
data_set.head()
```

```python
data_set.tail()
```

```python
data_set=data_set[ ["petal_length","petal_width"]]
data_set
```

```python
x=data_set['petal_length']
x
```

```python
y=data_set['petal_width']
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,color='red',marker='.')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)
x_test
```

```python
x_train
```

```python
y_test
```

```python
y_train
```

```python
import numpy as np
```

```python
x_train=np.array(x_train).reshape(-1)
x_train
```

```python
y_train=np.array(y_train).reshape(-1)
y_train
```

```python
x_test=np.array(x_test).reshape(-1)
x_test
```

```python
y_test=np.array(y_test).reshape(-1)
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
m=model.coef_
m
```

```python
c=model.intercept_
c
```

```python
y_train_pred=m*x_train+c
y_train_pred
```

```python
y_test_pred=m*x_test+c
y_test_pred
```

```python
plt.scatter(x_train,y_train,label='Data points',marker='.',color='r')
plt.plot(x_train,y_train_pred,color='g',label='RegressionLine')
plt.title('Linear Regression Line for Training Data')
plt.xlabel("petal length")
plt.ylabel("petal_width")
plt.legend()
plt.show()
```

```python
plt.title('Linear Regression Line for Testing Data')
plt.scatter(x_test, y_test, label='Data points', marker='.', color='r')
plt.plot(x_test, y_test_pred, label='Regression Line', color='g')
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.legend()
plt.show()

```
